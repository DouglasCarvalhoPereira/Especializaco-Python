# ENTRANDO NO SITE DE JOGOS DO DIA

```python
from selenium import webdriver
import time
import pandas as pd


link = "https://pbdatatrader.com.br/jogosdodia"
driver = webdriver.Chrome()
driver.get(link)


while len(driver.find_elements_by_tag_name('iframe')) == 0:
    time.sleep(2)
#Entrenado no primeiro iframe
iframe = driver.find_element_by_tag_name('iframe')
driver.switch_to.frame(iframe)
#Entrando no segundo iframe
iframe = driver.find_element_by_tag_name('iframe')
driver.switch_to.frame(iframe)
time.sleep(3)
#Primeira Tabela
lista_jogos = driver.find_element_by_class_name('innerContainer').text.split("\n")
columns = lista_jogos[:11]
valor_jogo = lista_jogos[11:]

dic = {}
for column in columns:
    dic[column] = []
    for jogo in valor_jogo[:20]:
        dic[column].append(jogo)
    valor_jogo = valor_jogo[20:]
    
tabela_df = pd.DataFrame.from_dict(dic)
time.sleep(1)

#Scrool na pagina e baixa as informações da planilha
for i in range(200):
    posicao_scroll = 350 * i
    driver.execute_script(f"document.getElementsByClassName('bodyCells')[0].scroll(0,{posicao_scroll})")
    
    #Baixa os textos da tabela, transforma em uma lista separadando os 11 primeiros elementos para colunas e os demais para valores
    lista_jogos = driver.find_element_by_class_name('innerContainer').text.split("\n")
    columns = lista_jogos[:11]
    valor_jogo = lista_jogos[11:]

    dic = {}
    for column in columns:
        dic[column] = []
        for jogo in valor_jogo[:20]:
            dic[column].append(jogo)
        valor_jogo = valor_jogo[20:]

    tabela_temporaria = pd.DataFrame.from_dict(dic)
    tabela_df = tabela_df.append(tabela_temporaria, ignore_index=True)

#Remover duplicatas
tabela_df = tabela_df.drop_duplicates()

#Armazenar tabela em Excel
tabela_df.to_excel("Tabela de Jogos.xlsx", index=False)

display(tabela_df)
```
