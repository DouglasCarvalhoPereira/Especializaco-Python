# Adicionando Colunas, Modificando Colunas e Valores

```python
import pandas as pd
#encoding='latin1', encoding='ISO-8859-1', encoding='utf-8' ou então encoding='cp1252'

vendas = pd.read_csv(r'dados/Contoso - Vendas  - 2017.csv', sep=';')
produtos = pd.read_csv(r'dados/Contoso - Cadastro Produtos.csv', sep=';')
lojas = pd.read_csv(r'dados/Contoso - Lojas.csv', sep=';')
clientes = pd.read_csv(r'dados/Contoso - Clientes.csv', sep=';')
#Excluindo Colunas
clientes = clientes[['ID Cliente', 'E-mail']]
produtos = produtos[['ID Produto', 'Categoria']]
lojas = lojas[['ID Loja', 'Nome da Loja']]

vendas = vendas.merge(produtos, on='ID Produto')
vendas = vendas.merge(lojas, on='ID Loja')
vendas = vendas.merge(clientes, on='ID Cliente').rename(columns={'E-mail': 'E-mail do Cliente'})

display(vendas)
```

## Acrescentando um COLUNA com mês, dia e ano de cada venda
Não só a data completa

```python
vendas.info()
```

```python
vendas['Data da Venda'] = pd.to_datetime(vendas['Data da Venda'], format="%d/%m/%Y")
vendas['Ano da Venda'] = vendas['Data da Venda'].dt.year
vendas['Mês da Venda'] = vendas['Data da Venda'].dt.month
vendas['Dia da Venda'] = vendas['Data da Venda'].dt.day


#É possível notar que o TYPE de Data da Venda agora é "datetime[ns]"
vendas.info()
display(vendas)
```
