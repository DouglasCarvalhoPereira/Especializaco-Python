# Mini projeto de análise de dados

### Importação de módulos e Arquivos


```python
import pandas as pd

funcionarios_df = pd.read_csv('dados/CadastroFuncionarios.csv', sep=';', decimal=',')
clientes_df = pd.read_csv('dados/CadastroClientes.csv', sep=';', decimal=',')
presta_servico_df = pd.read_excel('dados/BaseServiçosPrestados.xlsx', usecols = ['Codigo do Servico','ID Funcionário','ID Cliente','Tempo Total de Contrato (Meses)'])

#Excluir colunas
funcionarios_df = funcionarios_df.drop(['Estado Civil','Cargo'], axis=1)

display(funcionarios_df)
display(clientes_df)
display(presta_servico_df)
```

### 1 - Folha Slarial

```python
funcionarios_df['Salario Total'] = funcionarios_df['Salario Base'] + funcionarios_df['Impostos'] + funcionarios_df['Beneficios'] + funcionarios_df['VT'] + funcionarios_df['VR']
print("Total da folha salárial é: R$ {:,}".format(sum(funcionarios_df['Salario Total'])))
```

### 2 - Faturamento da Empresa

```python
faturamentos_df = presta_servico_df[['ID Cliente','Tempo Total de Contrato (Meses)']].merge(clientes_df[['ID Cliente','Valor Contrato Mensal']], on='ID Cliente')
faturamentos_df['Faturamento Total'] = faturamentos_df['Tempo Total de Contrato (Meses)'] * faturamentos_df['Valor Contrato Mensal']
display(faturamentos_df)
display("O faturamento Total é: R$ {:,}".format(sum(faturamentos_df['Faturamento Total'])))
```
