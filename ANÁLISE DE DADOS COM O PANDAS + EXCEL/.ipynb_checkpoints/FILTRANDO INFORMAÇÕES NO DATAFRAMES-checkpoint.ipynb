```python
import pandas as pd
#encoding='latin1', encoding='ISO-8859-1', encoding='utf-8' ou então encoding='cp1252'

vendas = pd.read_csv(r'dados/Contoso - Vendas  - 2017.csv', sep=';')
produtos = pd.read_csv(r'dados/Contoso - Cadastro Produtos.csv', sep=';')
lojas = pd.read_csv(r'dados/Contoso - Lojas.csv', sep=';')
clientes = pd.read_csv(r'dados/Contoso - Clientes.csv', sep=';')
#Excluindo Colunas
clientes = clientes[['ID Cliente', 'E-mail']]
produtos = produtos[['ID Produto', 'Categoria']]
lojas = lojas[['ID Loja', 'Nome da Loja']]

vendas = vendas.merge(produtos, on='ID Produto')
vendas = vendas.merge(lojas, on='ID Loja')
vendas = vendas.merge(clientes, on='ID Cliente').rename(columns={'E-mail': 'E-mail do Cliente'})

display(vendas)
```

# % De das vendas devolvidas

```python
quant_vendas = vendas['Quantidade Vendida'].sum()
quant_devolvidas = vendas['Quantidade Devolvida'].sum()

print("{:.2%}".format(quant_devolvidas / quant_vendas))
```

# Para saber a % de devoluções de uma loja específica ex. Loja C Europe Online

```python
vendas_europeonline = vendas[vendas['ID Loja'] == 306]
qt_devolvida = vendas_europeonline['Quantidade Devolvida'].sum()
qt_vendida = vendas_europeonline['Quantidade Vendida'].sum()
print("{:.2%}".format(qt_devolvida / qt_vendida))
```

# Entendendo o que está acontecendo

```python
loja306 = vendas['ID Loja'] == 306
print(loja306)
```

# Loja 306 sem devoluções

```python
loja306ZeroVendas = vendas[(vendas['ID Loja'] == 306) & (vendas['Quantidade Devolvida'] == 0)]
display(loja306ZeroVendas)
```
