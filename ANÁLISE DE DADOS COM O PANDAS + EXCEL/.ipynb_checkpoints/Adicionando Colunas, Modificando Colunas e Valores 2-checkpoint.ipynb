# Adicionando Colunas, Modificando Colunas e Valores 2

```python
import pandas as pd
#encoding='latin1', encoding='ISO-8859-1', encoding='utf-8' ou então encoding='cp1252'

vendas = pd.read_csv(r'dados/Contoso - Vendas  - 2017.csv', sep=';')
produtos = pd.read_csv(r'dados/Contoso - Cadastro Produtos.csv', sep=';')
lojas = pd.read_csv(r'dados/Contoso - Lojas.csv', sep=';')
clientes = pd.read_csv(r'dados/Contoso - Clientes.csv', sep=';')
#Excluindo Colunas
clientes = clientes[['ID Cliente', 'E-mail']]
produtos = produtos[['ID Produto', 'Categoria']]
lojas = lojas[['ID Loja', 'Nome da Loja']]

vendas = vendas.merge(produtos, on='ID Produto')
vendas = vendas.merge(lojas, on='ID Loja')
vendas = vendas.merge(clientes, on='ID Cliente').rename(columns={'E-mail': 'E-mail do Cliente'})

display(vendas)
```

# Modificação de valores específicos

## Métodos modificadores:

1. loc() Permite pegar uma lista de acordo com o índice dela. Ele dá errado caso não encontre o índice. 
   - 1.1. Podemos usar o loc() para acessar a linha e a coluna modificando diretamente um elemento.
        - loc[indiceLinha, indiceColuna]
        
2. iloc() Enxerga o dataframe como linha e colunas e consegue pegar o valor  com um número de linha e um número de coluna. Repara que ele não analisa o valor do indice e da coluna, apenas a posição importa.
    - Uso iloc[numLinha, numColuna]

```python
novo_produtos = pd.read_csv(r'dados/Contoso - Cadastro Produtos.csv', sep=';')
display(novo_produtos.head())
```

# Aumentar o preço do produto ID 873 para 2

```python
novo_produtos.loc['Contoso Wireless Laser Mouse E50 Grey', 'Preco Unitario'] = 23
display(novo_produtos.head())
```

```python
#Utilizando critério para atualizar e modificar valores
novo_produtos.loc[novo_produtos['ID Produto'] == 873, 'Preco Unitario'] = 23
display(novo_produtos.head())
```
