```python
import matplotlib.pyplot as plt
import numpy as np



vendas_meses = [1500, 1727, 1350, 999, 1050, 1027, 1022, 1500, 2000, 2362, 2100, 2762]
meses = ['jan', 'fev', 'mar', 'abr', 'mai', 'jun', 'jul', 'ago', 'set', 'out', 'nov', 'dez']

plt.plot(meses, vendas_meses)
plt.axis([0, 12, 0, max(vendas_meses) + 500])
plt.ylabel('Vendas')
plt.xlabel('Meses')
plt.show()

#plotar o gráfico da forma mais simples
```

```python
plt.bar(meses, vendas_meses)
plt.axis([0, 12, 0, max(vendas_meses) + 500])
plt.ylabel('Vendas')
plt.xlabel('Meses')
plt.show()
```

```python
plt.figure(1, figsize=(15, 3))
plt.subplot(1, 3, 1)
plt.plot(meses, vendas_meses)
plt.axis([0, 12, 0, max(vendas_meses) + 500])
plt.ylabel('Vendas')
plt.xlabel('Meses')
plt.subplot(1, 3, 2)
plt.bar(meses, vendas_meses)
plt.axis([0, 12, 0, max(vendas_meses) + 500])
plt.ylabel('Vendas')
plt.xlabel('Meses')
plt.subplot(1, 3, 3)
plt.plot(meses, vendas_meses)
plt.axis([0, 12, 0, max(vendas_meses) + 500])
plt.ylabel('Vendas')
plt.xlabel('Meses')
plt.show()
```
