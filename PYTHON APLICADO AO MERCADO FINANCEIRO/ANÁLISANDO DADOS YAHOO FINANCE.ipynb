### Vamos analisar o IBOV

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas_datareader.data as web

cotacao_ibov = web.DataReader('^BVSP', data_source='yahoo', start='2020-01-01', end='2020-12-31')
display(cotacao_ibov)
```

### Anaçisando o Gráfico

```python
cotacao_ibov['Adj Close'].plot(figsize=(15,5))
```

### Retorno do IBOV

```python
cotacao_ibov_inicio = cotacao_ibov['Adj Close'][-1] / cotacao_ibov['Adj Close'][0] - 1
print("Retorno IBOV: {:.2%}".format(cotacao_ibov_inicio))
```

### Analisando com Média Móvel

```python
cotacao_ibov['Adj Close'].plot(figsize=(15,5), label='IBOV') # Label = legenda
cotacao_ibov['Adj Close'].rolling(10).mean().plot(label='MM10D')
cotacao_ibov['Adj Close'].rolling(20).mean().plot(label='MM20D')
cotacao_ibov['Adj Close'].rolling(30).mean().plot(label='MM30D')

plt.legend() # Exibe a legenda
plt.show() # Sempre que for editar o gráfico usa plt.show()
```
